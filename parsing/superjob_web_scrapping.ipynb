### Шаг 1. 
#### По списку search_words ищем вакансии на странице https://russia.superjob.ru/vacancy/search/
#### и наполняем словарь data названием вакансии, ссылкой на нее и ключом isparsing

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time

# ВОТ ТУТ МОЖНО управлять через переменную max_parsing=? И ПАРСИНГ ПРОЙДЕТ НА МИНИМАЛКАХ )
max_parsing = 10

driver = webdriver.Chrome()
driver.get("https://russia.superjob.ru/vacancy/search/")
# список для поиска вакансий - их названия
search_words = [
    'аналитик данных'
# все позиции для поиска надо раскоментировать, если действительно надо спарсить все варианты
# для уменьшения времени работы кода я оставил один вариант названия вакансии 
    # ,'data analyst',
    # 'бизнес-аналитик', 
    # 'business analyst',
    # 'BI аналитик',
    # 'BI разработчик',
    # 'продуктовый аналитик',
    # 'маркетинговый аналитик',
    # 'финансовый аналитик',
    # 'веб-аналитик',
    # 'системный аналитик',
    # 'data scientist',
    # 'дата саентист',
    # 'ML инженер',
    # 'машинное обучение',
    # 'искусственный интеллект',
    # 'AI инженер',
    # 'computer vision',
    # 'NLP инженер',
    # 'глубокое обучение',
    # 'data engineer',
    # 'дата инженер',
    # 'big data',
    # 'ETL разработчик',
    # 'data architect',
    # 'data warehouse',
    # 'аналитик больших данных',
    # 'ведущий аналитик',
    # 'старший аналитик',
    # 'руководитель аналитики',
    # 'руководитель отдела данных',
    # 'главный аналитик',
    # 'head of data',
    # 'lead data scientist',
    # 'визуализация данных',
    # 'tableau',
    # 'power bi',
    # 'data quality',
    # 'качество данных',
    # 'дата сайентист',
    # 'инженер данных',
    # 'бизнес аналитик',
    # 'бизнес-аналитик',
    # 'мл инженер',
    # 'ai инженер'
]

data = {}
count = 1

# Быстро собираем данные )
# Я столкнулся с забавным наблюдением, что при запущенной конференции в Zoom 
# парсинг падал с ошибкой, не успевая за время sleep обработать страницу
# находим ссылки на вакансии, изначально их сильно больше и много не по теме
# поэтому пришлось разобраться с try except

for search_word in search_words:
    try:
        # Поиск и очистка поля ввода
        elem = driver.find_element(By.CSS_SELECTOR, "input[placeholder='Должность или компания']")
        elem.clear()  # Очищаем поле перед вводом
        elem.send_keys(search_word)
        elem.send_keys(Keys.RETURN)
        time.sleep(3) # часто даже увеличение ожидания не помогло
        
        keywords = search_word.split()
        
        # Используем более специфичный селектор для вакансий на SuperJob
        vak_elements = driver.find_elements(By.CSS_SELECTOR, "a[href*='/vakansii/'], a[href*='/vacancy/']")
        time.sleep(2)
        
        # Собираем данные с обработкой исключений
        vak_data = []
        for vak in vak_elements:
            try:
                href = vak.get_attribute('href')
                text = vak.text 
                if href and text:
                    vak_data.append((text.strip(), href))
            except Exception as e:
                print(f"Ошибка при получении данных элемента: {e}")
                continue
        
        # Обрабатываем данные
        for text, href in vak_data:
            if any(keyword in text.lower() for keyword in keywords):
                data[count] = {
                    'name': text,
                    'salary': '',
                    'currency': '',
                    'experience': '',
                    'location': '',
                    'skills': '',
                    'extra_data': '',
                    'link': href,
                    'isparsing': 0 # идея в том, что когда я пойду парсить все линки вакансий и прервусь - можно продолжить со значений "0"
                    # потому что при парсинге я ставлю здесь "1"
                }
                count += 1
                if (count > max_parsing):
                    break
        
        # Возвращаемся на страницу поиска, при поиске одного названия вакансии - по сути это конец
        driver.get("https://russia.superjob.ru/vacancy/search/")
        time.sleep(2)
        
    except Exception as e:
        # print(f"Ошибка при обработке запроса '{search_word}': {e}")
        # В случае ошибки перезагружаем страницу поиска
        driver.get("https://russia.superjob.ru/vacancy/search/")
        time.sleep(2)

driver.quit()
```

### Шаг 2. 
#### Проходим по всем 'link' из словаря наполняем описание необходимыми данными

```python
import json

# тут получлась чуть монструозая функция, но как говорится "как умею" )))
def extract_salary_from_json_ld(driver):
    """Извлекает зарплату из structured data (JSON-LD)"""
    # Ищем script tag с JSON-LD
    script_elements = driver.find_elements(By.CSS_SELECTOR, "script[type='application/ld+json']")
    
    # Инициализируем переменные для местоположения
    address_locality = None
    address_country = None
    address_region = None
    address = None
    
    # Пытаемся найти опыт работы на странице
    experience = None
    skills = []
    extra_data = []
    
    try:
        # experience_element = driver.find_element(By.CSS_SELECTOR, "span._3Kf6x._3KGkE._2LsIK")
        experience_element = driver.find_element(By.XPATH, "/html/body/div[1]/div/div/div[1]/div[4]/div/div[1]/div/div[1]/div/div[4]/div[1]/div[1]/div[1]/div[3]")
        experience = experience_element.text
    except:
        # Если элемент не найден, оставляем experience как None
        pass

    
    # поиск навыков или описание вакансии, если навыки явно не указаны
    # далее при обогащении/чистке данных создадим словарь с ключевыми скилами и по ним поищем уже в описании дополнив ключ skills
    try:
        # Ищем блок по заголовку "Профессиональные навыки"
        skills_header = driver.find_element(By.XPATH, "//span[text()='Профессиональные навыки']")
        skills_section = skills_header.find_element(By.XPATH, "./../../..")
        
        # Ищем все навыки в этом блоке
        skill_elements = skills_section.find_elements(By.CSS_SELECTOR, "span.GDfng._3KGkE.NrB9b._2LsIK._2UP26")
        
        for element in skill_elements:
            text = element.text.strip()
            if text and text not in skills:
                skills.append(text)
                
    except:
        # Если нет такого раздела, ищем по альтернативному классу
        try:
            # тут происходит переодическая замена CSS_SELECTOR, не придумал как делать на постоянку - обнолвение от 12.12.2025
            alternative_elements = driver.find_elements(By.CSS_SELECTOR, "span.mrLsm _23xe3 _2_ehB _3H8Cg _3Tn5h _1rYf7")
            for element in alternative_elements:
                text = element.text.strip()
                if text and text not in skills:
                    extra_data.append(text)
        except:
            pass
    
    for script in script_elements:
        try:
            json_data = json.loads(script.get_attribute('innerHTML'))
            
            # Organization
            # addressLocality addressCountry addressRegion
            # "jobLocation":{"@type":"Place","address":{"@type":"PostalAddress","addressLocality":"Старый Крым","addressCountry":"RU","addressRegion":"Республика Крым"}}
            if 'jobLocation' in json_data:
                job_location = json_data['jobLocation']
                
                if 'address' in job_location:
                    address = job_location['address']
                    address_locality = address.get('addressLocality')
                    address_country = address.get('addressCountry')
                    address_region = address.get('addressRegion')
        
            # Проверяем наличие зарплаты в structured data
            if 'baseSalary' in json_data:
                salary_info = json_data['baseSalary']
                if 'value' in salary_info:
                    value = salary_info['value']
                    min_salary = value.get('minValue')
                    
                    if min_salary:
                        return {
                            'salary': min_salary,
                            'currency': salary_info.get('currency', 'RUB'),
                            'location': address_locality,                    # город или др населенный пункт
                            'location2': address_country,                    # страна
                            'location3': address_region,                     # регион
                            'location_full': address,                        # полный набор
                            'skills': ', '.join(skill.replace('\n', ', ').strip() for skill in skills) if skills else None,  # Стэк / ключевые навыки
                            'extra_data': ', '.join(ed.replace('\n', ', ').strip() for ed in extra_data) if extra_data else None, # фактически описание вакансии
                            'experience': experience                         # опыт работы
                        }
        except json.JSONDecodeError:
            continue
    
    # Если зарплата не найдена, возвращаем 0
    return {
        'salary': 0,
        'currency': 'RUB',
        'location': address_locality,                                   # город или др населенный пункт
        'location2': address_country,                                   # страна
        'location3': address_region,                                    # регион
        'location_full': address,                                       # полный набор
        'skills': ', '.join(skill.replace('\n', ', ').strip() for skill in skills) if skills else None,  # Стэк / ключевые навыки
        'extra_data': ', '.join(ed.replace('\n', ', ').strip() for ed in extra_data) if extra_data else None, # фактически описание вакансии
        'experience': experience                                        # опыт работы
    }

# пора использовать эту функцию
count = 1
data2 = {}
for d in data.values():
    if d.get('isparsing') == 0: # вот этот заветный нолик )) - просто несколько раз рвался парсинг - пришлось выкручиваться
        driver = webdriver.Chrome()
        driver.get(d.get('link'))
        salary_data = extract_salary_from_json_ld(driver)
        # вспомогательный словарь
        data2[count] = {
                    'salary': salary_data.get('salary'), 
                    'currency': salary_data.get('currency'),
                    'experience': salary_data.get('experience'), 
                    'location': salary_data.get('location'),
                    'skills': salary_data.get('skills'),
                    'extra_data': salary_data.get('extra_data')
                }
        d.update({'isparsing': 1})
        time.sleep(3)
        driver.quit()
    count += 1

    if (count > max_parsing):
        break
```

```python
# дополним основной словарь по числовому ключу 
for key in data:
    if key in data2:
        data[key].update(data2[key])
```
